```python
soup = BeautifulSoup(html_code, 'html.parser')
```

```python
li_elements = soup.find_all("li")
```

```python
StoreName=[]
StoreAddress=[]
City=[]
Phone=[]
Website=[]

for li in li_elements:
  div_stores = li.find("div")
  phone_element = li.find('a', href=lambda href: href and href.startswith('tel:'))
  phone = phone_element['href'][4:] if phone_element else "Not Available"
  Phone.append(phone)
  #print("Phone:", phone)

  website_element = li.find('a', href=lambda href: href and href.startswith("http") and not href.startswith("https://www.google.com"))
  website = website_element['href'][4:] if website_element else "Not Available"
  Website.append(website)
  #print("Website:", website)

  if div_stores:
    store_name_element = div_stores.find('p').find('strong')
    store_name = store_name_element.text if store_name_element else "Not Available"
    StoreName.append(store_name)
    #print("Store Name:", store_name)

    #Extract the store address
    store_address_element = div_stores.find('p').find('span', class_='wpsl-street')
    store_address = store_address_element.text if store_address_element else "Not Available"
    StoreAddress.append(store_address)
    #print("Store Address:", store_address)

    city_element = div_stores.find_all('span')[1]
    city = city_element.text if city_element else "Not Available"
    City.append(city)
    #print("City:", city)
```
