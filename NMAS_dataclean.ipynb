```python
import csv
import pandas as pd
from google.colab import files
```

```python
def filter_empty_strings(value):
    return value != ' '
```

```python
column=[]
with open("NMAS_PDF_csv.csv", 'r') as file:
  csvreader = csv.reader(file)
  for row in csvreader:
    for cell in row:
      cells = cell.split('\n')
      columns = list(filter(filter_empty_strings, cells))
      column.append(columns)
```

```python
df = pd.DataFrame(column)
```

```python
df.to_excel("NMAS_dataclean.xlsx", index=False)
files.download('NMAS_dataclean.xlsx')
```
